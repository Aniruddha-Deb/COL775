```python
import torch
import torchvision
from torchvision.datasets import MNIST
import torch.nn as nn
import torch.optim as optim
from tqdm.auto import tqdm
import matplotlib.pyplot as plt

from torch.utils.data import Dataset, DataLoader
```

This notebook implements the original Variational Autoencoder paper (Kingma and Wellig, 2014)

```python
device = torch.device('cpu')
```

```python
class MNIST_LinearVAE(nn.Module):
    
    def __init__(self, z_dim=2):
        
        super().__init__()
        
        self.encoder = nn.Sequential(
            nn.Linear(784,128),
            nn.ReLU(),
            nn.Dropout(p=0.1),
            nn.Linear(128,128),
            nn.ReLU(),
            nn.Dropout(p=0.1),
        )
        
        self.enc_mean = nn.Linear(128,z_dim)
        self.enc_var = nn.Linear(128,z_dim)
        
        self.decoder = nn.Sequential(
            nn.Linear(z_dim,128),
            nn.ReLU(),
            nn.Dropout(p=0.1),
            nn.Linear(128,128),
            nn.ReLU(),
            nn.Dropout(p=0.1),
            nn.Linear(128,784),
            nn.Sigmoid()
        )
        
        self.loss_fn = nn.BCELoss(reduction='none')
        
    def forward(self, x):
        
        enc_features = self.encoder(x)
        z_mu = self.enc_mean(enc_features)
        z_logvar = self.enc_var(enc_features)

        # reparameterization trick, l=1
        z = z_mu + torch.exp(0.5*z_logvar)*torch.randn_like(z_logvar)
        
        x_hat = self.decoder(z)
                
        l1 = self.loss_fn(x_hat,x).sum(axis=1).mean()
        l2 = (0.5*(1 + z_logvar - z_mu**2 - torch.exp(z_logvar))).sum(axis=1).mean()
        
        return l1-l2
```

```python
transform = torchvision.transforms.Compose([torchvision.transforms.ToTensor()])
train_ds = MNIST('.', transform=transform)
test_ds = MNIST('.', transform=transform, train=False)
```

```python
model = MNIST_LinearVAE(z_dim=2).to(device)
```

```python
train_dl = DataLoader(train_ds, batch_size=100, num_workers=0, shuffle=True)
```

```python
optimizer = optim.Adam(model.parameters(), lr=1e-4)
scheduler = optim.lr_scheduler.StepLR(optimizer, 2, gamma=0.2)
```

```python
losses = []
for epoch in range(5):
    
    model.train()
    n_batches = 1000
    i = 0
    train_done=False
    for batch in tqdm(train_dl):

        i += 1
        imgs = batch[0].reshape((100,28*28)).to(device)
        optimizer.zero_grad()
        loss = model(imgs)
        if (loss < 0):
            train_done=True
            break
        loss.backward()
        optimizer.step()
        losses.append(loss.detach())

        if (i > n_batches):
            break
    scheduler.step()
    
    if train_done:
        break

plt.plot(losses)
```

```python
fig, axes = plt.subplots(21, 21, figsize=(12,12))
model.eval()
for i in range(21):
    for j in range(21):
        y = 3*((i/10) - 1)
        x = 3*((j/10) - 1)
        axes[i][j].set_axis_off()
        with torch.no_grad():
            axes[i][j].imshow(1-model.decoder(torch.tensor([[x,y]]).float()).detach().reshape(28,28).numpy(), cmap='gray')
```
