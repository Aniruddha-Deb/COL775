```python
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
```

```python
train_curves = pkl.load(open('pt2_train_curves.pkl', 'rb'))
val_curves = pkl.load(open('pt2_val_curves.pkl', 'rb'))
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for name, curve in train_curves.items():
    ax.plot(range(1,len(curve)+1), curve, alpha=0.9, label=f'{name}_train')
ax.set_prop_cycle(None)
for name, curve in val_curves.items():
    ax.plot(range(1,len(curve)+1), curve, '--', alpha=0.6, label=f'{name}_val')
ax.legend()
```
