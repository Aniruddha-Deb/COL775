```python
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
```

```python
train_curves = pkl.load(open('pt2_train_curves.pkl', 'rb'))
val_curves = pkl.load(open('pt2_val_curves.pkl', 'rb'))
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for name, curve in train_curves.items():
    ax.plot(range(1,len(curve)+1), curve, alpha=0.9, label=f'{name}_train')
ax.set_prop_cycle(None)
for name, curve in val_curves.items():
    ax.plot(range(1,len(curve)+1), curve, '--', alpha=0.6, label=f'{name}_val')
ax.legend()

ax.set_title('ResNet - different normalization schemes')
ax.set_xlabel('epoch')
ax.set_ylabel('loss')
fig.savefig('report/q2_part3.pdf')
```

```python
fdata = pkl.load(open('feature_quantile_data.pkl', 'rb'))
```

```python
quantiles = list(zip(*fdata))
epoch_range = range(1,len(quantiles[0])+1)
plt.plot(epoch_range, quantiles[0], label='1st')
plt.plot(epoch_range, quantiles[1], label='20th')
plt.plot(epoch_range, quantiles[2], label='80th')
plt.plot(epoch_range, quantiles[3], label='99th')
plt.title('Feature Quantiles of first validation example')
plt.xlabel('epoch')
plt.legend()
plt.ylabel('value')
plt.savefig('report/q2_part7.pdf')
```
